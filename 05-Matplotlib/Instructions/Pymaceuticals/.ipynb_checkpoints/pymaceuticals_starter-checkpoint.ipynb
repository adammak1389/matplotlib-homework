## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_data_combined_df = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="outer")

# Display the data table for preview
mouse_data_combined_df
```

```python
# Checking the number of mice.
mouse_data_combined_df["Mouse ID"].nunique()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
mouse_data_combined_df[mouse_data_combined_df.duplicated(['Mouse ID', 'Timepoint'], keep=False)]

# Optional: Get all the data for the duplicate mouse ID. 

duplicated_mouse_data_combined_df = mouse_data_combined_df[mouse_data_combined_df.duplicated(['Mouse ID', 'Timepoint'], keep=False)]
duplicated_mouse_data_combined_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_mouse_data_df = mouse_data_combined_df.drop_duplicates(subset=['Mouse ID', 'Timepoint'], keep='first')
cleaned_mouse_data_df
```

```python
# Checking the number of mice in the clean DataFrame.
cleaned_mouse_data_df["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
regimen_summary_df = cleaned_mouse_data_df.groupby('Drug Regimen').agg(
             regimen_mean=('Tumor Volume (mm3)', 'mean'),
             regimen_median=('Tumor Volume (mm3)', 'median'),
             regimen_variance=('Tumor Volume (mm3)', 'var'),
             regimen_std_deviation=('Tumor Volume (mm3)', 'std'),
             regimen_sem=('Tumor Volume (mm3)', 'sem'))
regimen_summary_df
# This method produces everything in a single groupby function
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pandas.
total_mice_treatments_df = pd.DataFrame(cleaned_mouse_data_df.groupby("Drug Regimen")["Mouse ID"].nunique())

total_mice_treatments_df
```

```python
total_mice_treatments_df.rename(columns={"Mouse ID": "Total Mice"})
```

```python
total_mice_treatments_df.plot(kind="bar", figsize=(20,5))

# set a title for the chart
plt.title("Total Mice Per Treatment")
plt.ylabel("Count of Mice")
plt.grid()
plt.show()
```

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pyplot.

mice_counts = [25, 25, 25, 25, 25, 25, 25, 25, 24, 25]
x_axis = np.arange(len(mice_counts))

plt.bar(x_axis, mice_counts, color='r', alpha=1, align="center")
plt.title("Total Mice Per Treatment")
plt.xlabel("Drug Regimen")
plt.ylabel("Count of Mice")
# Tell matplotlib where we would like to place each of our x axis headers
plt.xticks(x_axis, ["Capomulin", "Ceftamin", "Infubinol", "Ketapril", "Naftisol", "Placebo", "Propriva", "Ramicane", "Stelasyn", "Zoniferol" ], rotation=90)
# set the x limits of the current chart
plt.xlim(-0.75, x_axis[-1] + 0.75)
# set the y limits of the current chart
plt.ylim(0, max(mice_counts)*1.1)
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin


# Start by getting the last (greatest) timepoint for each mouse

greatest_timepoint_df = pd.DataFrame(cleaned_mouse_data_df.groupby("Mouse ID")["Timepoint"].max())
greatest_timepoint_df
```

```python
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
timepoint_data_combined_df = pd.merge(cleaned_mouse_data_df, greatest_timepoint_df, on=["Mouse ID","Timepoint"], how="inner")
timepoint_data_combined_df
```

```python
timepoint_data_combined_df = timepoint_data_combined_df[~timepoint_data_combined_df['Drug Regimen'].isin(['Ketapril', 'Naftisol', 'Placebo', 'Propriva', 'Stelasyn', 'Zoniferol'])]
timepoint_data_combined_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
treatments_tumor_vol = []

for vols in timepoint_data_combined_df["Tumor Volume (mm3)"]:
    treatments_tumor_vol.append(vols)
treatments_tumor_vol

for treatments in treatments_list:
    treatments_list.append([timepoint_data_combined_df["Mouse ID"], timepoint_data_combined_df["Tumor Volume (mm3)"]])
treatments_list
```

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_data_combined_df = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="outer")

# Display the data table for preview
mouse_data_combined_df
```

```python
# Checking the number of mice.
mouse_data_combined_df["Mouse ID"].nunique()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
mouse_data_combined_df[mouse_data_combined_df.duplicated(['Mouse ID', 'Timepoint'], keep=False)]

# Optional: Get all the data for the duplicate mouse ID. 

duplicated_mouse_data_combined_df = mouse_data_combined_df[mouse_data_combined_df.duplicated(['Mouse ID', 'Timepoint'], keep=False)]
duplicated_mouse_data_combined_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_mouse_data_df = mouse_data_combined_df.drop_duplicates(subset=['Mouse ID', 'Timepoint'], keep='first')
cleaned_mouse_data_df
```

```python
# Checking the number of mice in the clean DataFrame.
cleaned_mouse_data_df["Mouse ID"].nunique()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

correlation = st.pearsonr(capomulin_mouse_weight_filtered_df["mouse_weight"],capomulin_mouse_weight_filtered_df["avg_tumor_vol"])
print(f"The correlation between both factors is {round(correlation[0],2)}")
```

```python
(slope, intercept, rvalue, pvalue, stderr) = linregress(capomulin_mouse_weight_filtered_df["mouse_weight"], capomulin_mouse_weight_filtered_df["avg_tumor_vol"])
print(slope,intercept,rvalue,pvalue,stderr)

regress_values = capomulin_mouse_weight_filtered_df["mouse_weight"] * slope + intercept
```

```python
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(capomulin_mouse_weight_filtered_df["mouse_weight"], capomulin_mouse_weight_filtered_df["avg_tumor_vol"])

print(f"The Linear Regression equation is: {line_eq}")
plt.plot(capomulin_mouse_weight_filtered_df["mouse_weight"], regress_values, "r-")
plt.annotate(line_eq, (5,5), fontsize=10, color="red")
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```
